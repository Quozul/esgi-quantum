```python
from qiskit import QuantumCircuit
circuit = QuantumCircuit(3)
circuit.draw('mpl')
```

```python
circuit.h(0)  # h = porte de Hadamard = état de superposition d'un qubit
circuit.draw('mpl')
```

```python
circuit.cx(0, 1)  # cx = CNOT = Control Not = Intrication
circuit.draw('mpl')
```

```python
# Simulateur quantique
# Aer : le module des simulateurs

from qiskit import Aer

# Création d'un moteur de calcul
backend = Aer.get_backend('statevector_simulator')
```

```python
# Lancer le calcul
job = backend.run(circuit)
# Afficher le résultat
result = job.result()
output_state = result.get_statevector(circuit, decimals=3)
output_state
```

```python
# Afficher un histogramme
from qiskit.visualization import plot_state_city
plot_state_city(output_state)
```
